# Описание проекта GPN-CUP по Data Science 2022
## Цели и задачи проекта

Во вселенной Гарри Поттера, жители волшебного мира подсмотрели у маглов способы повышения заработка для своих баров, пабов и трактиров. Они стали проводить различные акции и кампании, пытаясь увеличить объём покупаемых напитков, после чего каждый трактирщик начал вести записи о том, что, кому и сколько было продано.

Эта практика стала настолько распространённой, что в 1998 году сотрудники Министерства Магии, собирая налоги, решили подсчитать, насколько честны трактирщики и сколько налогов могут принести подобные акции в будущем. Даже вторая магическая война с тем «Кого нельзя называть», когда продажи сильно менялись, не стала помехой. Таким образом, Министерство Магии хочет:

+ Посчитать по имеющимся данным, как каждая акция, во время её проведения, повлияла на объём приобретённых напитков или иначе говоря, какой дополнительный объём напитков был приобретён посетителями трактиров, в результате проведения акций в волшебном мире.
+ Построить модель для предсказания:
    + объёма продаж напитков без учёта промоакций, чтобы проверять трактирщиков на честность.
    + дополнительного объёма продаж напитков от промоакций, чтобы узнать сколько дополнительных налогов принесут эти акции.

Обсудим отдельно задачи предсказания:
+ Задачу предсказания без учёта промоакций можно решить двумя способами:
    + как обычную задачу регрессии, но тогда проверять трактирщиков на честность можно будет только по факту, так как будут необходимы записи о реализации напитков для формирования признаков.
    + как задачу регрессии для временных рядов, тогда предсказания будут известны заранее, но возможно они будут хуже, так как признаки будут генерироваться искусственно.
+ Задачу предсказания от промоакций будем решать, как задачу регрессии для временных рядов, так как Министерство Магии хочет заранее знать, сколько дополнительных налогов они получат в будущем.

## Описание данных

Все данные собраны в 6 датасетах:

__Таблица *daily_volume*__ — дневные объёмы реализации напитков.
+ `city` — город заведения
+ `id` — идентификатор заведения
+ `product` — название напитка
+ `date` — дата реализации
+ `volume_litr` — объём реализации напитка

__Таблица *effect*__ — эффект от промоакций.
+ `date` — дата расчёта эффекта
+ `product` — название напитка
+ `city` — город промоакции
+ `volume_litr` — объём эффекта в дополнительных литрах

__Таблица *sales*__ — реализация напитков клиентам из магических орденов.
+ `date` — дата реализации
+ `drink` — является ли напиток алкогольным
+ `city` — город
+ `segment` — группа клиентов
+ `volume_litr` — объём в литрах

__Таблица *attack_probability*__ — вероятность нападения на разные города.
+ `city` — город
+ `date` — дата
+ `attack probability` — вероятность нападения (Орденом Феникса или Пожирателями смерти)

__Таблица *promo*__ — описание промоакций.
+ `Название` — наименование акции
+ `Начало` — дата начала акции
+ `Конец` — дата окончания акции
+ `Dummy-столбцы` — присутствие акции в определённом городе
+ `Dummy-столбцы` — является ли напиток алкогольным
+ `Затраты на акцию, галлеонов` — затраты на акцию в галлеонах

__Таблица *drinks*__ — виды напитков.
+ `Напиток` — название напитка
+ `Тип` — является ли напиток алкогольным

__Также, в данных встречается следующая терминология:__
+ `Галлеон` — золотая монета денежной системы во вселенной Гарри Поттера.
+ `Чешуя дракона` — разновидность эля (пива) во вселенной Гарри Поттера.
+ `Орден Феникса` — организация, основанная Альбусом Дамблдором для борьбы с лордом Волан-де-Мортом и Пожирателями смерти.
+ `Пожиратели смерти` — группа тёмных волшебников, последователей лорда Волан-де-Морта, сражавшихся во второй магической войне.

__Также стоит отметить,__ что столбец `volume_litr` таблицы `daily_volume`, включает в себя значения столбца `volume_litr` таблицы `sales`.

### Категоризация напитков
Напишем функцию __define_type_drink()__, определяющую является ли напиток алкогольным и создадим новый признак __type_drink__ для каждой из таблиц __daily_volume__ и __effect__.

```python
%%time
# Время выполнения ~30 секунд
def define_type_drink(row):
    alcoholic_beverage = ['Сливочное пиво', 'Чешуя дракона', 'Огненный виски', 'Смородиновый ром']
    line = row['product']
    if line in alcoholic_beverage:
        return 'Алкогольный'
    return 'Безалкогольный'


daily_volume['type_drink'] = daily_volume.apply(define_type_drink, axis=1)
effect['type_drink'] = effect.apply(define_type_drink, axis=1)
```

Также проверим, есть ли такие безопасные города, вероятность нападения на которые всегда равна нулю.

```python
safe_cities = []

for i in unique_cities:
    # chance_attack - это уникальные значения вероятностей нападения на i-ый город
    chance_attack = attack_probability[attack_probability['city'] == i]['attack_probability'].unique()
    # Если в chance_attack лежит только одно значение и это ноль, то город безопасный
    if len(chance_attack) == 1:
        if chance_attack == 0:
            safe_cities.append(i)
            
if len(safe_cities) == 0:
    print('Безопасных городов нет')
else:
    print(f'Безопасные города: {safe_cities}')
```

# Задача №1 — Влияние акций на дополнительный объём продаж
Чтобы сравнить влияние различных акций на дополнительные объёмы реализации напитков, построим столбчатую диаграмму с общими объёмами дополнительных продаж напитков по каждой промоакции. Для этого сделаем следующее:
+ Напишем функцию __count_promo()__ для расчёта действующего количества промоакций на каждую строку таблицы __effect__.
+ Напишем функцию __promo_impact()__ для расчёта суммарного дополнительного объёма продаж с каждой промоакции. Данная функция использует результат работы функции __count_promo()__.
+ Далее уже можно будет строить столбчатую диаграмму по обновлённой таблице __promo__.

Для комфортного чтения функций, выведем по 2 строчки таблиц __effect__ и __promo__.

```python
effect.head(2)
```

```python
promo.head(2)
```

```python
%%time
# Время выполнения ~40 секунд
def count_promo(df):  # Функция ждёт таблицу effect в качестве аргумента для параметра df
    # Так как несколько акций могут проходить в один и тот же день, то
    # создадим счётчик акций на каждую запись в effect
    count = 0
    # Создадим список городов для проверки условия по городу
    cities = ['Кромер', 'Париж', 'Бат', 'Лондон', 'Салем', 'Коукворт',
              'Авимор', 'Даффтаун', 'Абердин', 'Нью-Йорк', 'Литтл Уингинг']
    # Будем последовательно перебирать каждую акцию из таблицы promo
    for i in range(promo.shape[0]):
        # Условие №1 - Сверяем дату записи в effect с датой проведения акции
        if df['date'] >= promo['Начало'][i] and df['date'] <= promo['Конец'][i]:
            # Если дата совпадает, то переходим к проверке следующего условия
            # Условие №2 - Проверяем город
            for city in cities:
                if df['city'] == city and promo[city][i] == 1:
                    # Если город совпадает, то переходим к проверке следующего условия
                    # Условие №3 - Проверяем тип алкогольности напитка
                    if ((df['type_drink'] == 'Алкогольный' and promo['Алкогольное'][i] == 1) or 
                        (df['type_drink'] == 'Безалкогольный' and promo['Безалкогольное'][i] == 1)):
                        # Если тип алкогольности напитка совпадает, то увеличиваем значение count на 1
                        count += 1
                        # Далее используем конструкцию break, так как нам нужно только одно совпадение по городу,
                        # а цикл перебора по городам продолжит свою работу. Поэтому прервём его, после первого совпадения
                        break
    # count - это количество промоакций, проходящих в одно время для каждой строки таблицы effect
    return count


# Запишем новый признак в столбец count_promo
effect['count_promo'] = effect.apply(count_promo, axis=1)
# Посмотрим на промежуточный результат
effect.head(3)
```

```python
%%time
# Время выполнения ~40 секунд
def promo_impact(promo):  # Функция ждёт таблицу promo в качестве аргумента для параметра promo
    # Cоздадим счётчик дополнительных объёмов продаж напитков от промоакций
    total_volume = 0
    # Создадим список городов для проверки условия по городу
    cities = ['Кромер', 'Париж', 'Бат', 'Лондон', 'Салем', 'Коукворт',
              'Авимор', 'Даффтаун', 'Абердин', 'Нью-Йорк', 'Литтл Уингинг']
    # Будем последовательно перебирать каждую строку из таблицы effect
    for i in range(effect.shape[0]):
        # Условие №1 - Сверяем дату записи в effect с датой проведения акции
        if effect['date'][i] >= promo['Начало'] and effect['date'][i] <= promo['Конец']:
            # Если дата совпадает, то переходим к проверке следующего условия
            # Условие №2 - Проверяем город
            for city in cities:
                if effect['city'][i] == city and promo[city] == 1:
                    # Если город совпадает, то переходим к проверке следующего условия
                    # Условие №3 - Проверяем тип алкогольности напитка
                    if ((effect['type_drink'][i] == 'Алкогольный' and promo['Алкогольное'] == 1) or 
                        (effect['type_drink'][i] == 'Безалкогольный' and promo['Безалкогольное'] == 1)):
                        # Если тип алкогольности напитка совпадает, то идём дальше. Так как в один день может
                        # быть несколько акций, то и дополнительные продажи будут распределены на несколько акций,
                        # поэтому будем делить volume_litr на равные доли между проходящими в этот день акциями.
                        total_volume += effect['volume_litr'][i] / effect['count_promo'][i]
                        # Далее используем конструкцию break, так как нам нужно только одно совпадение по городу,
                        # а цикл перебора по городам продолжит свою работу. Поэтому прервём его, после первого совпадения
                        break
    return total_volume


# Запишем новый признак в столбец additional_volume
promo['additional_volume'] = promo.apply(promo_impact, axis=1)
```

Далее, на основе проведённого анализа данных и сделанных на нём выводов, попробуем добавить в __daily_volume__ следующие признаки:

### Добавление признаков entries_day и uniq_cities
+ __entries_day__ — количество записей сделанных трактирщиками за каждый день.
+ __uniq_cities__ — количество уникальных городов за каждый день.

```python
# Создадим временную таблицу с группировкой по дате и городу
temp_table = (daily_volume.groupby(['date', 'city'])
                          .agg(liters_city=('volume_litr', 'sum'), entries_city=('volume_litr', 'count'))
                          .reset_index())

# Сгруппируем temp_table по дате и поместим в новый датафрейм drink_sales_volume
# Далее уже в drink_sales_volume будем добавлять все остальные признаки
drink_sales_volume = (temp_table.groupby('date')
                                .agg(liters_day=('liters_city', 'sum'),
                                     entries_day=('entries_city', 'sum'),
                                     uniq_cities=('city', 'count'))
                                .reset_index())
# Посмотрим, что получилось
drink_sales_volume.head(15)
```

### Добавление признака mean_promo
Добавим среднее количество промоакций на каждый день. У нас уже есть функция __count_promo()__ для расчёта действующего количества промоакций на каждую строку таблицы. Её можно применить к таблице __daily_volume__ с помощью метода __apply()__, но так как таблица __daily_volume__ содержит 3.7+ млн записей, то сначала сгруппируем её по дате, городу и типу алкогольности напитка. Это уменьшит исходный датасет в 200+ раз и время выполнения функции до разумного, а на значение среднего количества промоакций никак не повлияет.

```python
# Создадим временную таблицу temp_table_2, чтобы не менять исходный daily_volume
temp_table_2 = daily_volume.groupby(['date', 'city', 'type_drink']).agg(volume_litr=('volume_litr', 'sum')).reset_index()
# Посмотрим, что получилось
temp_table_2.head(3)
```

Значения столбца __mean_promo__ получились вещественными числами, но по своему смыслу этот признак ближе к целому числу. Округлим его и приведём к типу __int__.

### Добавление признака chance_attack
Вероятность нападения на город может влиять на посещаемость заведений и соответственно на объёмы продаваемых напитков. Рассчитаем с какой обобщённой вероятностью может произойти нападение в конкретный день на любой из городов.
+ У нас есть вероятность события __A__ — нападение на город в какой-то день. Тогда противоположное ему событие __Ā = 1 - А__, то есть вероятность того, что на город не нападут. Также __противоположное событие__ ещё можно охарактеризовать как: __все другие исходы__.
+ Допустим в i-ый день есть 3 города на которые могут напасть с вероятностями __A1__, __A2__ и __A3__. Чтобы узнать вероятность возникновения одновременно всех трёх событий, нужно перемножить их вероятности.
+ Тогда вероятность того, что ни на один город не нападут равна __B = (1 - А1) х (1 - А2) х (1 - А3)__.
+ Тогда событие противоположное __B__ — это событие __B̅ = 1 - (1 - А1) х (1 - А2) х (1 - А3)__.
+ Поскольку __B__ — это вероятность что ни на один город не нападут, то __противоположное событие B̅__ — это __все остальные исходы__, то есть вероятность, что нападут на все города сразу или только на несколько или только на один.
+ Наш признак __chance_attack__ будет хранить именно вероятность события __B̅__.

Перед тем, как создавать признак, проверим следующее:
+ количество уникальных городов в каждом дне
+ нет ли пропущенных дат в столбце

```python
# Создадим временную таблицу temp_table_3, чтобы не менять исходный attack_probability
temp_table_3 = (attack_probability.groupby(['date'])
                                  .agg(uniq_cities=('city', 'count'))
                                  .reset_index())

temp_table_3 = temp_table_3.sort_values('date')
temp_table_3
```

```python
temp_table_3['uniq_cities'].unique()
```

Получается, что каждая строка таблицы __attack_probability__ — это одна уникальная дата и один уникальный город. Теперь проверим даты так же, как проверяли у таблицы __drink_sales_volume__.

```python
temp_table_3['date'][334] - temp_table_3['date'][0]
```

Всё хорошо. Пропущенных значений в датах нет. Теперь начнём формировать признак.

```python
# Запишем вероятность ненападения для каждой строки
attack_probability = attack_probability.sort_values('date')
attack_probability['non_aggression'] = 1 - attack_probability['attack_probability']
attack_probability.head(3)
```

Далее можно сгруппировать __attack_probability__ по дате и в качестве агрегирующей функции использовать умножение в столбце __non_aggression__, но такой агрегации нет, поэтому напишем свою функцию и используем её через метод __apply()__.

```python
def calculation_chance_attack(df):  # Функция ждёт таблицу temp_table_3 в качестве аргумента для параметра df
    # В line построчно будет записываться столбец date из temp_table_3
    line = df['date']
    # Создадим счётчик для перемножения шансов ненападения
    count = 1
    for i in range(len(attack_probability)):
        if attack_probability['date'][i] == line:
            count *= attack_probability['non_aggression'][i]
    # Возвращаем все перемноженные шансы для даты line из temp_table_3
    return count


# Запишем промежуточный признак в столбец full_non_aggression
temp_table_3['full_non_aggression'] = temp_table_3.apply(calculation_chance_attack, axis=1)
# Финальный обобщённый шанс нападения запишем в столбец chance_attack
temp_table_3['chance_attack'] = 1 - temp_table_3['full_non_aggression']
# Удалим лишние столбцы из temp_table_3
temp_table_3 = temp_table_3.drop(['uniq_cities', 'full_non_aggression'], axis = 1)
# Посмотрим что получилось
temp_table_3.head(5)
```

Теперь, чтобы добавить признак __chance_attack__ в таблицу __drink_sales_volume__, используем объединение таблиц методом __merge__.

```python
# Будем объединять типом слияния left
drink_sales_volume = drink_sales_volume.merge(temp_table_3, on='date', how='left')
# Посмотрим что получилось
drink_sales_volume
```

Так как, временной диапазон таблицы __drink_sales_volume__ шире, чем у таблицы __temp_table_3__ на целый год, то первые и последние строки столбца __chance_attack__ заполнились пропусками. Будем считать, что вероятность нападения равна нулю для тех дней, которые не учтены в таблице __attack_probability__ и затем соответственно в таблице __temp_table_3__.

```python
# Заполним пропуски нулями
drink_sales_volume['chance_attack'] = drink_sales_volume['chance_attack'].fillna(0)
# Посмотрим что получилось
drink_sales_volume.head(5)
```

## Создание признаков для задач №3 и №4
Напишем функцию __make_features()__ для создания признаков. Так как у нас два разных целевых признака, то скользящее среднее и отстающие значения тоже будут разными. Поэтому для каждой из наших задач, запишем результат работы функции __make_features()__ в отдельную переменную.

```python
def make_features(dt, rolling_mean_size, max_lag, name_target = 'liters_day'):
    # Будем возвращать копию, чтобы не менять оригинал
    data = dt.copy()
    
    # Добавим календарные признаки
    # Добавим порядковый день в месяце
    data['month_day'] = data.index.day
    # Добавим месяц
    data['month'] = data.index.month
    # Добавим год
    data['year'] = data.index.year
    # Добавим день недели, где 0 - это Понедельник, а 6 - это Воскресенье
    data['week_day'] = data.index.dayofweek
    # Добавим порядковый день в году
    data['year_day'] = data.index.dayofyear
    
    # Добавим скользящее среднее
    data['rolling_mean'] = data[name_target].shift().rolling(rolling_mean_size).mean()
    # Добавим отстающие значения
    for lag in range(1, max_lag + 1):
        data[f'lag_{lag}'] = data[name_target].shift(lag)
    # Удалим строки с пропусками, которые появятся после добавления скользящего среднего и отстающих значений
    data = data.dropna()
    
    return data
```

+ Обучающая выборка — 70%
+ Валидационная выборка — 20%
+ Тестовая выборка — 10%

При создании переменных, используем приставку __add__ для выборок, отвечающих за предсказание дополнительного объёма продаж от промоакций.

```python
# Для предсказания объёма продаж
features = realized_volume.drop('liters_day', axis=1)
target = realized_volume['liters_day']

features_train, features_valid, target_train, target_valid = train_test_split(
    features,
    target,
    shuffle=False,
    test_size=0.3,
    random_state=12345)

features_valid, features_test, target_valid, target_test = train_test_split(
    features_valid,
    target_valid,
    shuffle=False,
    test_size=0.333,
    random_state=12345)


# Для предсказания дополнительного объёма продаж
features_add = additional_volume.drop('additional_liters_day', axis=1)
target_add = additional_volume['additional_liters_day']

features_train_add, features_valid_add, target_train_add, target_valid_add = train_test_split(
    features_add,
    target_add,
    shuffle=False,
    test_size=0.3,
    random_state=12345)

features_valid_add, features_test_add, target_valid_add, target_test_add = train_test_split(
    features_valid_add,
    target_valid_add,
    shuffle=False,
    test_size=0.333,
    random_state=12345)
```

## Обучение модели Случайного леса для задачи №3

```python
# Счётчики модели, метрики, количества деревьев и глубины дерева для метрики MAE
best_model_for_mae = None
best_score_for_mae = 1_000_000
best_est_for_mae = 0
best_depth_for_mae = 0

# Счётчики модели, метрики, количества деревьев и глубины дерева для метрики RMSE
best_model_for_rmse = None
best_score_for_rmse = 1_000_000
best_est_for_rmse = 0
best_depth_for_rmse = 0

for est in range(1, 11):
    for depth in range(1, 15):
        model = RandomForestRegressor(random_state=12345, n_estimators=est, max_depth=depth)
        model.fit(features_train, target_train)
        predicted_valid = model.predict(features_valid)
        mae = round(mean_absolute_error(target_valid, predicted_valid))
        rmse = round(mean_squared_error(target_valid, predicted_valid)**0.5)
        
        # Подбор лучших показателей для метрики MAE
        if mae < best_score_for_mae:
            best_model_for_mae = model
            best_score_for_mae = mae
            best_est_for_mae = est
            best_depth_for_mae = depth
        
        # Подбор лучших показателей для метрики RMSE
        if rmse < best_score_for_rmse:
            best_model_for_rmse = model
            best_score_for_rmse = rmse
            best_est_for_rmse = est
            best_depth_for_rmse = depth

            
print('Модель - Случайный лес')
print()
print('Значение метрик для предсказания объёма продаж:')
print(f'Лучшее MAE {best_score_for_mae} при количестве деревьев {best_est_for_mae} и глубине дерева {best_depth_for_mae}')
print(f'Лучшее RMSE {best_score_for_rmse} при количестве деревьев {best_est_for_rmse} и глубине дерева {best_depth_for_rmse}')
```
